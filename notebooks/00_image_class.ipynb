```python
import sys
sys.path.insert(0, '../src')

import os.path
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from aspire.image import Image

_im = np.load('monuments.npy')
print(_im.shape)
```

```python
def show(im):
    plt.figure(figsize=(20,10))
    columns = 5
    for i in range(im.shape[-1]):
        plt.subplot(im.shape[-1] / columns + 1, columns, i + 1)
        plt.imshow(im[:, :, i], cmap='gray')
```

```python
# Use the following line to re-initialize the Image in any cell
show(im)
```

```python
shifts = np.array([100, 30])  # A single x/y shift, or shifts corresponding to each image
show(im.shift(shifts))
```

```python
show(im.downsample(80))
```

```python
from aspire.utils.filters import CTFFilter

# pixel_size/defous_u/defocus_v in Angstrom, voltage in kV
filter = CTFFilter(pixel_size=1, voltage=100, defocus_u=1500, defocus_v=2000)
show(im.filter(filter))

filter = CTFFilter(pixel_size=1, voltage=1000, defocus_u=1500, defocus_v=2000)
show(im.filter(filter))
```
